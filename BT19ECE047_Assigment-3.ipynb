# Task 3

```python
c_img = np.asarray(c_img)
r_img = np.asarray(r_img)
```

```python
# AND gate
and_img = np.zeros((200, 200))
for i in range(200):
    for j in range(200):
        if(c_img[i,j] == 255 and r_img[i,j] == 255):
            and_img[i,j] = 255

plt.imshow(and_img, cmap='gray')
```

```python
# OR gate
or_img = np.zeros((200, 200))
for i in range(200):
    for j in range(200):
        if(c_img[i,j] == 255 or r_img[i,j] == 255):
            or_img[i,j] = 255

plt.imshow(or_img, cmap='gray')
```

```python
#NAND Gate
nand_img = np.zeros((200, 200))
for i in range(200):
    for j in range(200):
        if(not(c_img[i,j] == 255 and r_img[i,j] == 255)):
            nand_img[i,j] = 255

plt.imshow(nand_img, cmap='gray')
```

```python
# NOR gate
nor_img = np.zeros((200, 200))
for i in range(200):
    for j in range(200):
        if(not(c_img[i,j] == 255 or r_img[i,j] == 255)):
            nor_img[i,j] = 255

plt.imshow(nor_img, cmap='gray')
```

```python
# XOR gate
xor_img = np.zeros((200, 200))
for i in range(200):
    for j in range(200):
        if(c_img[i,j] != r_img[i,j]):
            xor_img[i,j] = 255

plt.imshow(xor_img, cmap='gray')
```

```python
#NOT gate
not_img = np.zeros((200, 200))
for i in range(200):
    for j in range(200):
        if(xor_img[i,j] == 0):
            not_img[i,j] = 255

plt.imshow(not_img, cmap='gray')
```
